```python

def order2_line(point1, point2, point3):
    x = np.array([point1[0], point2[0], point3[0]])
    y = np.array([point1[1], point2[1], point3[1]])

    A = np.array([
        [0, 0, 1],
        [1, 1, 1],
        [4, 2, 1],
    ])

    return np.linalg.solve(A, x), np.linalg.solve(A, y)
```

```python
order2_line([0,1], [2,2], [8,5])
```

```python
order2_line([1,1], [2,3], [3,9])
```

```python
order2_line([1,1], [5, 3], [9, 17])
```
