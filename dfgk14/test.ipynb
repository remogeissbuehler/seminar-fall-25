```python
class Test:
    def __init__(self, x):
        self.x = x
        self.a = "a"

    def get_a(self):
        return self.a
```

```python
def get_x(self, mult=1):
    return self.x * mult
```

```python
test = Test(42)
test.get_a()
```

```python
Test.get_x = get_x
```

```python
test.get_x()
```

```python
test.get_x(mult=3)
```
